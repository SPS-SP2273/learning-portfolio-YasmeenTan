```python
import imageio

# Read the images and store them in a list
images = []
for i in range(1, 18):
    filename = f'golgi-movie_robert_frame_{i:02d}_gray.jpg'
    images.append(imageio.imread(filename))

# Calculate the total fluorescence signal of each image
fluorescence = []
for image in images:
    total_fluorescence = np.sum(image)
    fluorescence.append(total_fluorescence)

# Create a time axis
time = np.arange(1, 18)
    
# Plot the intensities according to images
plt.plot(time, fluorescence)
plt.xlabel('Frame Number')
plt.ylabel('Total fluorescence signal')
plt.show()
```
